```python
scale = 4

paper_size = (595, 842)
qr_size = (54, 54)

xs = [
    20.5,
    89,
    157.5,
    226,
    314.5,
    383,
    451.5,
    520
]
ys = [
    67,
    176,
    285,
    394,
    503,
    612,
    721
]

paper_size = [int(x * scale) for x in paper_size]
qr_size = [int(x * scale) for x in qr_size]
qr_map = [[int(x * scale), int(y * scale)] for x in xs for y in ys]
```

```python
import random
import string
import qrcode

rand_chars = string.ascii_letters + string.digits

used_ids = set()

def make_qr():
    while True:
        id = ''.join(random.choice(rand_chars) for _ in range(5))
        
        if id not in used_ids:
            break

    used_ids.add(id)

    qr = qrcode.QRCode(
        border=0,
    )
    qr.add_data(id)
    qr.make(fit=True)
    img = qr.make_image(back_color="transparent")
    img = img.resize(qr_size, resample=0)

    return id, img


def make_page():
    page = Image.new('RGBA', paper_size, (255, 255, 255, 255))
    for x, y in qr_map:
        id, img = make_qr()
        page.paste(img, (x, y), img)

    return page
```

```python
from PIL import Image
from IPython.display import display
import os

used_ids = set()
count = 300
i = 0

os.makedirs('../QR', exist_ok=True)

while len(used_ids) < 300:
    page = make_page()
    page.save(f'../QR/{i}.png')
    display(page)
    i += 1
```
